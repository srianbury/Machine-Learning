```python
import pandas
from bs4 import BeautifulSoup as bs
import re
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
data = pandas.read_csv("imdb_reviews.tsv", delimiter="\t")
```

```python
def clean_text_and_remove_stopwords(review):
    list_of_words_sans_stopwords = []
    rev_soup = bs(review, "lxml") #use beautiful soup to remove html markup
    review_with_letters_only = re.sub("[^a-zA-Z]", " ", rev_soup.get_text()) #substitute any character that isnt a-z or A-Z with " "
    review_with_letters_only_and_all_lower_case = review_with_letters_only.lower() #make all words lowercase
    list_of_words =  review_with_letters_only_and_all_lower_case.split() #turn the string into a list of words
    for word in list_of_words:
        if word not in stopwords.words("english"):
            list_of_words_sans_stopwords.append(word)
    sentence_sans_stopwords = " ".join(list_of_words_sans_stopwords)
    return sentence_sans_stopwords
```

```python
NUMBER_OF_REVIEWS = 25
```

```python
list_of_cleaned_reviews = []
for index in range(len(data["review"][0:NUMBER_OF_REVIEWS])):
    list_of_cleaned_reviews.append(clean_text_and_remove_stopwords(data["review"][index]))

for review in list_of_cleaned_reviews:
    print(review + "\n")
```

```python
(train_reviews, test_reviews, train_target, test_target) = \
train_test_split(list_of_cleaned_reviews, data["sentiment"][0:NUMBER_OF_REVIEWS],\
test_size = 0.2)
```

```python
#bag of words with 5000 most common words
vectorizer = CountVectorizer(analyzer='word', max_features = 5000)

#find the right 5000 words
vectorizer.fit(train_reviews)

#look at which words it found
print(vectorizer.get_feature_names())
```

```python
#transform review strings into the rows of word counts
#also convert it to a numpy array so we can later feed it to a machine learning algorithm

train_word_columns = vectorizer.transform(train_reviews).toarray()
test_word_columns = vectorizer.transform(test_reviews).toarray()

#look at what train_words_columns looks like
print(train_word_columns)
```

```python
model = MultinomialNB()
model.fit(train_word_columns, train_target)
predictions = model.predict(test_word_columns)
print(accuracy_score(predictions, test_target))
```
